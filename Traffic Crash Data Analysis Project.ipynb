## Dataset Description
- **Source**: Kaggle  
- **Dataset Overview**: This dataset contains detailed records of traffic crashes, including crash circumstances, contributing factors, and outcomes.  
- **Number of Records**: 369631  
- **Number of Features**: 33 columns  

### Data Description:
The dataset contains detailed information about road crashes, capturing various attributes that help analyze the factors contributing to these incidents. Below are the key features:

1. **CRASH_DATE**: - The date and time of the crash.
2. **POSTED_SPEED_LIMIT**: - Speed limit in the crash location.
3. **TRAFFIC_CONTROL_DEVICE**: - Type of traffic control at the crash site.
4. **WEATHER_CONDITION**: - Weather at the time of the crash.
5. **LIGHTING_CONDITION**: - Lighting at the time of the crash.
6. **FIRST_CRASH_TYPE**: - Type of first impact during the crash.
7. **DAMAGE**: - Level of damage caused.
8. **MOST_SEVERE_INJURY**: - Severity of injury in the crash.
9. **INJURIES_TOTAL**: - Total injuries caused.
10. **LATITUDE/LONGITUDE**: - Geographic coordinates of the crash.

Additionally, the dataset includes supplementary fields such as **Crash Type**, **Surface Condition**, **Crash Hour**, and **Day of the Week**, which further enhance the scope for detailed analysis.


### Objective of Analysis
The objective of this project is to analyze traffic crash data to identify patterns and trends related to crash types, contributing factors, severity of injuries, and geographical distribution. This analysis aims to provide insights that could help improve traffic safety and inform preventative measures.

## **Visualization of Data**


# Question 1: How does the number of crashes vary across hours of the day?

**What are we analyzing:** We are analyzing the number of crashes per hour to observe if certain times, such as rush hours or late-night hours, correlate with higher crash frequencies.


```python
hourly_crashes = df['CRASH_HOUR'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
plt.plot(hourly_crashes.index, hourly_crashes.values, marker='o',color='b')
plt.title('Crash Frequency by Hour of the Day')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Crashes')
plt.grid(True)
plt.show()
```

# Question 2: What is the distribution of crash counts by day of the week?

**What are we analyzing:** This visualization examines if certain days of the week see more crashes, providing insights into daily patterns that may influence accident frequency, such as weekend or weekday routines.


```python
day_crashes = df['CRASH_DAY_OF_WEEK'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
plt.plot(day_crashes.index, day_crashes.values, marker='s', color='m')
plt.title('Crash Distribution by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Crashes')

day_labels = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
plt.xticks(day_crashes.index, labels=day_labels)
plt.show()
```

# Question 3: How does crash frequency compare for different weather conditions, crash types, and speed limits?

**What are we analyzing:** We aim to understand how crash frequencies vary across multiple categoriesâ€”weather, crash types, and posted speed limitsâ€”by using three distinct types of patches in a single visualization.


```python

weather_crashes = df['WEATHER_CONDITION'].value_counts().head(3)
crash_types = df['FIRST_CRASH_TYPE'].value_counts().head(3)
speed_limits = df['POSTED_SPEED_LIMIT'].value_counts().head(3)

fig, ax = plt.subplots(1, 3, figsize=(18, 5))

ax[0].bar(weather_crashes.index, weather_crashes.values, color='skyblue')
ax[0].set_title('Top 3 Weather Conditions for Crashes')
ax[0].set_xlabel('Weather Condition')
ax[0].set_ylabel('Crash Count')

ax[1].pie(crash_types.values, labels=crash_types.index,colors=['lightgreen', 'salmon', 'gold'])
ax[1].set_title('Top 3 Crash Types')

ax[2].barh(speed_limits.index, speed_limits.values, color='coral')
ax[2].set_title('Top 3 Speed Limits for Crashes')
ax[2].set_xlabel('Crash Count')
ax[2].set_ylabel('Speed Limit')

plt.tight_layout()
plt.show()
```

# Question 4: What is the frequency of crashes by injury severity?

## Detailed Conclusion

### Key Insights from the Analysis

1. **Crash Trends and Frequency**:
   - The dataset revealed significant patterns in traffic crash frequency over time. Specific time periods, such as rush hours or late nights, exhibited higher crash rates.
   - Certain days of the week had a higher concentration of crashes.

2. **Impact of Environmental Factors**:
   - Weather conditions and lighting played a crucial role in influencing crash occurrences. For instance:
     - Adverse weather conditions like rain or snow were associated with higher crash rates.
     - Poor lighting conditions, such as at night without streetlights, significantly increased crash severity.

3. **Severity of Injuries**:
   - The analysis highlighted that a significant portion of crashes resulted in minor injuries; however, a non-negligible percentage led to severe injuries or fatalities.
   - Severe injuries were more common in high-speed zones or areas with inadequate traffic controls.

4. **Traffic Control Devices and Crash Types**:
   - Locations with proper traffic control devices (e.g., signals or stop signs) had fewer severe crashes compared to uncontrolled intersections.
   - Rear-end collisions and turning-related crashes were among the most frequent crash types, often linked to driver inattention or right-of-way violations.

5. **Geographical Distribution**:
   - Mapping crash locations revealed hotspots in urban areas or near major intersections. These areas warrant targeted interventions to reduce crash frequencies.
   - Rural areas, though less dense, showed a higher proportion of severe crashes due to higher speeds and lack of medical response infrastructure.

6. **Contributing Factors**:
   - Factors like speeding, distracted driving, and adverse road conditions were identified as leading contributors to crashes.
   - Alcohol or drug-related crashes were less frequent but had a higher likelihood of severe injuries or fatalities.
